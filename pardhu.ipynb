```python
# ==========================================================
# RESNET18 + CAPSULE NETWORK
# FULL METRICS + VISUALIZATIONS (JUPYTER + RTX 3060)
# ==========================================================

import os, glob, json, math
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
from torchvision.transforms import RandAugment

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

from tqdm import tqdm
from torch.amp import autocast, GradScaler

torch.backends.cudnn.benchmark = True

# ==========================================================
# CONFIG
# ==========================================================
class CFG:
    CLASS_NAMES = ['IGA0','IGA1','IGA2','IGA3','IGA4']
    NUM_CLASSES = 5

    IMG = 224
    BATCH = 16
    EPOCHS = 25

    LR = 3e-4
    MIN_LR = 1e-6
    WARMUP = 2
    WD = 1e-4

    PRIMARY_DIM = 8
    PRIMARY_CHANNELS = 32
    DIGIT_DIM = 16
    ROUTING_ITERS = 3

    M_PLUS = 0.9
    M_MINUS = 0.1
    LAMBDA = 0.5
    MARGIN_SCALE = 5.0

    CE_WARMUP_EPOCHS = 5

    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cfg = CFG()
print("Using device:", cfg.DEVICE)

# ==========================================================
# HELPERS
# ==========================================================
def squash(x, dim=-1, eps=1e-7):
    n = torch.norm(x, dim=dim, keepdim=True)
    return (n**2/(1+n**2))*(x/(n+eps))

def cosine_lr(epoch):
    if epoch < cfg.WARMUP:
        return cfg.LR*(epoch+1)/cfg.WARMUP
    t = (epoch-cfg.WARMUP)/max(1,cfg.EPOCHS-cfg.WARMUP)
    return cfg.MIN_LR + 0.5*(cfg.LR-cfg.MIN_LR)*(1+math.cos(math.pi*t))

def ce_alpha(epoch):
    return 1.0 if epoch < cfg.CE_WARMUP_EPOCHS else 0.0

# ==========================================================
# DATASET
# ==========================================================
class AcneDataset(Dataset):
    def __init__(self, root, split, transform):
        self.paths, self.labels = [], []
        self.transform = transform

        for i,c in enumerate(cfg.CLASS_NAMES):
            d = os.path.join(root, split, c)
            if not os.path.exists(d): continue
            for ext in ("*.jpg","*.png","*.jpeg","*.JPG","*.PNG"):
                for p in glob.glob(os.path.join(d,ext)):
                    self.paths.append(p)
                    self.labels.append(i)

        print(f"✓ {split} set: {len(self.paths)} images")

    def __len__(self): return len(self.paths)

    def __getitem__(self, i):
        img = Image.open(self.paths[i]).convert("RGB")
        return self.transform(img), torch.tensor(self.labels[i],dtype=torch.long)

train_tf = transforms.Compose([
    transforms.Resize((cfg.IMG,cfg.IMG)),
    RandAugment(2,9),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

val_tf = transforms.Compose([
    transforms.Resize((cfg.IMG,cfg.IMG)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

# ==========================================================
# MODEL
# ==========================================================
class PrimaryCaps(nn.Module):
    def __init__(self,in_ch):
        super().__init__()
        self.conv = nn.Conv2d(in_ch,cfg.PRIMARY_CHANNELS*cfg.PRIMARY_DIM,1)

    def forward(self,x):
        x=self.conv(x)
        B,_,H,W=x.shape
        x=x.view(B,cfg.PRIMARY_CHANNELS,cfg.PRIMARY_DIM,H*W)
        x=x.permute(0,1,3,2).contiguous().view(B,-1,cfg.PRIMARY_DIM)
        return squash(x)

class DigitCaps(nn.Module):
    def __init__(self,in_caps):
        super().__init__()
        self.W = nn.Parameter(0.01*torch.randn(1,in_caps,cfg.NUM_CLASSES,cfg.DIGIT_DIM,cfg.PRIMARY_DIM))

    def forward(self,x):
        B=x.size(0)
        x_hat=torch.matmul(self.W.expand(B,-1,-1,-1,-1),x.unsqueeze(2).unsqueeze(-1)).squeeze(-1)
        b=torch.zeros(B,x_hat.size(1),cfg.NUM_CLASSES,1,device=x.device)
        for _ in range(cfg.ROUTING_ITERS):
            c=F.softmax(b,dim=2)
            s=(c*x_hat).sum(dim=1)
            v=squash(s)
            b=b+(x_hat*v.unsqueeze(1)).sum(dim=-1,keepdim=True)
        return v

class ResNetCaps(nn.Module):
    def __init__(self):
        super().__init__()
        base=models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        self.backbone=nn.Sequential(*list(base.children())[:-2])
        self.primary=PrimaryCaps(512)
        with torch.no_grad():
            n=self.primary(self.backbone(torch.zeros(1,3,224,224))).size(1)
        self.digit=DigitCaps(n)

    def forward(self,x):
        f=self.backbone(x)
        p=self.primary(f)
        d=self.digit(p)
        l=torch.norm(d,dim=2)
        return l,l.argmax(1)

# ==========================================================
# LOSS
# ==========================================================
class CapsuleLoss(nn.Module):
    def __init__(self,counts):
        super().__init__()
        counts=np.asarray(counts,dtype=np.float32)
        counts[counts<1]=1
        w=counts.sum()/counts
        w=(w/w.mean()).clip(0.6,2.0)
        self.register_buffer("w",torch.tensor(w,dtype=torch.float32))
        self.ce=nn.CrossEntropyLoss(weight=self.w)

    def forward(self,l,y,alpha):
        t=F.one_hot(y,cfg.NUM_CLASSES).float()
        p=F.relu(cfg.M_PLUS-l)**2
        a=F.relu(l-cfg.M_MINUS)**2
        margin=(t*p+cfg.LAMBDA*(1-t)*a).sum(dim=1).mean()
        return cfg.MARGIN_SCALE*margin + alpha*self.ce(l,y)

# ==========================================================
# TRAIN + VALIDATE
# ==========================================================
def train(root):
    os.makedirs("evaluation_results",exist_ok=True)

    train_ds=AcneDataset(root,"train",train_tf)
    val_ds=AcneDataset(root,"valid",val_tf)
    test_ds=AcneDataset(root,"test",val_tf)

    counts=np.bincount(train_ds.labels,minlength=cfg.NUM_CLASSES)
    sampler=WeightedRandomSampler(1/(counts[np.array(train_ds.labels)]+1e-6),
                                  len(train_ds.labels),replacement=True)

    train_loader=DataLoader(train_ds,batch_size=cfg.BATCH,sampler=sampler,num_workers=0)
    val_loader=DataLoader(val_ds,batch_size=cfg.BATCH,num_workers=0)
    test_loader=DataLoader(test_ds,batch_size=cfg.BATCH,num_workers=0)

    model=ResNetCaps().to(cfg.DEVICE)
    loss_fn=CapsuleLoss(counts).to(cfg.DEVICE)
    opt=optim.AdamW(model.parameters(),lr=cfg.LR,weight_decay=cfg.WD)
    scaler=GradScaler(enabled=(cfg.DEVICE.type=="cuda"))

    history={"train_loss":[],"val_loss":[],"train_acc":[],"val_acc":[]}

    for epoch in range(cfg.EPOCHS):
        model.train()
        lr=cosine_lr(epoch)
        for g in opt.param_groups: g["lr"]=lr
        alpha=float(ce_alpha(epoch))

        tr_loss, y_true, y_pred = 0, [], []

        for x,y in tqdm(train_loader,desc=f"Epoch {epoch+1}",ncols=80):
            x,y=x.to(cfg.DEVICE),y.to(cfg.DEVICE)
            opt.zero_grad(set_to_none=True)
            with autocast(device_type=cfg.DEVICE.type,enabled=cfg.DEVICE.type=="cuda"):
                l,p=model(x)
                loss=loss_fn(l,y,alpha)
            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()

            tr_loss+=loss.item()
            y_true.extend(y.cpu().numpy())
            y_pred.extend(p.cpu().numpy())

        train_acc=accuracy_score(y_true,y_pred)*100

        # ---------- VALIDATION ----------
        model.eval()
        v_loss, yt, yp = 0, [], []
        with torch.no_grad():
            for x,y in val_loader:
                x,y=x.to(cfg.DEVICE),y.to(cfg.DEVICE)
                l,p=model(x)
                loss=loss_fn(l,y,0.0)
                v_loss+=loss.item()
                yt.extend(y.cpu().numpy())
                yp.extend(p.cpu().numpy())

        val_acc=accuracy_score(yt,yp)*100

        history["train_loss"].append(tr_loss/len(train_loader))
        history["val_loss"].append(v_loss/len(val_loader))
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        print(f"Epoch {epoch+1}: "
              f"Train Acc {train_acc:.2f}% | Val Acc {val_acc:.2f}%")

    # ================= FINAL METRICS =================
    cm=confusion_matrix(yt,yp)
    report=classification_report(yt,yp,target_names=cfg.CLASS_NAMES)
    precision=precision_score(yt,yp,average="macro")
    recall=recall_score(yt,yp,average="macro")
    f1=f1_score(yt,yp,average="macro")

    print("\nPrecision:",precision)
    print("Recall:",recall)
    print("F1:",f1)
    print("\n",report)

    # ================= PLOTS =================
    plt.figure(); plt.plot(history["train_loss"],label="Train Loss")
    plt.plot(history["val_loss"],label="Val Loss"); plt.legend(); plt.title("Loss")
    plt.savefig("evaluation_results/loss.png"); plt.show()

    plt.figure(); plt.plot(history["train_acc"],label="Train Acc")
    plt.plot(history["val_acc"],label="Val Acc"); plt.legend(); plt.title("Accuracy")
    plt.savefig("evaluation_results/accuracy.png"); plt.show()

    plt.figure(figsize=(6,5))
    plt.imshow(cm,cmap="Blues"); plt.title("Confusion Matrix")
    plt.colorbar(); plt.xticks(range(cfg.NUM_CLASSES),cfg.CLASS_NAMES,rotation=45)
    plt.yticks(range(cfg.NUM_CLASSES),cfg.CLASS_NAMES)
    plt.savefig("evaluation_results/confusion_matrix.png"); plt.show()

    return model

# ==========================================================
# RUN
# ==========================================================
train(r"D:\Mini")
```

```python
# ==========================================================
# SWIN TRANSFORMER (PRETRAINED) - JUPYTER + RTX 3060
# Multi-class Image Classification with Full Metrics
# ==========================================================

import os, glob, json
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms

import timm
from tqdm import tqdm
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)

from torch.amp import autocast, GradScaler

torch.backends.cudnn.benchmark = True

# ==========================================================
# CONFIG
# ==========================================================
class CFG:
    DATA_ROOT = r"D:\Mini"     # <<< CHANGE IF NEEDED
    CLASS_NAMES = ['IGA0','IGA1','IGA2','IGA3','IGA4']
    NUM_CLASSES = 5

    IMG_SIZE = 224
    BATCH = 16        # Try 32 if memory allows
    EPOCHS = 20
    LR = 3e-4
    WEIGHT_DECAY = 1e-4

    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cfg = CFG()
print("Using device:", cfg.DEVICE)

# ==========================================================
# DATASET
# ==========================================================
class AcneDataset(Dataset):
    def __init__(self, root, split, transform):
        self.paths, self.labels = [], []
        self.transform = transform

        for idx, cname in enumerate(cfg.CLASS_NAMES):
            d = os.path.join(root, split, cname)
            if not os.path.exists(d): 
                continue
            for ext in ("*.jpg","*.png","*.jpeg","*.JPG","*.PNG"):
                for p in glob.glob(os.path.join(d, ext)):
                    self.paths.append(p)
                    self.labels.append(idx)

        print(f"✓ {split} set: {len(self.paths)} images")

    def __len__(self): 
        return len(self.paths)

    def __getitem__(self, i):
        img = Image.open(self.paths[i]).convert("RGB")
        return self.transform(img), torch.tensor(self.labels[i], dtype=torch.long)

# ==========================================================
# TRANSFORMS
# ==========================================================
train_tf = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(0.1,0.1,0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
])

val_tf = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],
                         [0.229,0.224,0.225])
])

# ==========================================================
# MODEL (SWIN TRANSFORMER)
# ==========================================================
def build_model():
    model = timm.create_model(
        "swin_tiny_patch4_window7_224",
        pretrained=True,
        num_classes=cfg.NUM_CLASSES
    )
    return model.to(cfg.DEVICE)

# ==========================================================
# TRAIN / EVAL
# ==========================================================
@torch.no_grad()
def evaluate(model, loader):
    model.eval()
    y_true, y_pred = [], []
    loss_sum = 0.0

    for x, y in loader:
        x, y = x.to(cfg.DEVICE), y.to(cfg.DEVICE)
        logits = model(x)
        loss = criterion(logits, y)
        loss_sum += loss.item()

        preds = logits.argmax(1)
        y_true.extend(y.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

    acc = accuracy_score(y_true, y_pred) * 100
    return loss_sum / len(loader), acc, np.array(y_true), np.array(y_pred)

# ==========================================================
# MAIN TRAIN FUNCTION
# ==========================================================
def train_swin():
    os.makedirs("evaluation_results", exist_ok=True)

    train_ds = AcneDataset(cfg.DATA_ROOT, "train", train_tf)
    val_ds   = AcneDataset(cfg.DATA_ROOT, "valid", val_tf)
    test_ds  = AcneDataset(cfg.DATA_ROOT, "test",  val_tf)

    # ---- Class balancing
    counts = np.bincount(train_ds.labels, minlength=cfg.NUM_CLASSES)
    weights = 1.0 / (counts[np.array(train_ds.labels)] + 1e-6)
    sampler = WeightedRandomSampler(weights, len(weights), replacement=True)

    train_loader = DataLoader(train_ds, batch_size=cfg.BATCH,
                              sampler=sampler, num_workers=0)
    val_loader   = DataLoader(val_ds, batch_size=cfg.BATCH,
                              shuffle=False, num_workers=0)
    test_loader  = DataLoader(test_ds, batch_size=cfg.BATCH,
                              shuffle=False, num_workers=0)

    model = build_model()
    print("✓ Swin Transformer loaded")

    global criterion
    criterion = nn.CrossEntropyLoss().to(cfg.DEVICE)
    optimizer = optim.AdamW(model.parameters(), lr=cfg.LR,
                            weight_decay=cfg.WEIGHT_DECAY)

    scaler = GradScaler(enabled=(cfg.DEVICE.type == "cuda"))

    history = {"train_loss":[], "val_loss":[],
               "train_acc":[], "val_acc":[]}

    best_val = 0.0
    best_path = "evaluation_results/best_swin.pth"

    # ================= TRAIN LOOP =================
    for epoch in range(cfg.EPOCHS):
        model.train()
        train_loss = 0.0
        y_true, y_pred = [], []

        for x, y in tqdm(train_loader, desc=f"Epoch {epoch+1}/{cfg.EPOCHS}"):
            x, y = x.to(cfg.DEVICE), y.to(cfg.DEVICE)
            optimizer.zero_grad(set_to_none=True)

            with autocast(device_type=cfg.DEVICE.type,
                          enabled=(cfg.DEVICE.type=="cuda")):
                logits = model(x)
                loss = criterion(logits, y)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            train_loss += loss.item()
            y_true.extend(y.cpu().numpy())
            y_pred.extend(logits.argmax(1).cpu().numpy())

        train_acc = accuracy_score(y_true, y_pred) * 100
        val_loss, val_acc, _, _ = evaluate(model, val_loader)

        history["train_loss"].append(train_loss/len(train_loader))
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        print(f"Epoch {epoch+1}: "
              f"Train Acc={train_acc:.2f}% | Val Acc={val_acc:.2f}%")

        if val_acc > best_val:
            best_val = val_acc
            torch.save(model.state_dict(), best_path)
            print("✓ Best model saved")

    # ================= TEST + METRICS =================
    model.load_state_dict(torch.load(best_path, map_location=cfg.DEVICE))
    test_loss, test_acc, y_true, y_pred = evaluate(model, test_loader)

    precision = precision_score(y_true, y_pred, average="macro", zero_division=0)
    recall    = recall_score(y_true, y_pred, average="macro", zero_division=0)
    f1        = f1_score(y_true, y_pred, average="macro", zero_division=0)
    cm        = confusion_matrix(y_true, y_pred)
    report    = classification_report(
                    y_true, y_pred, target_names=cfg.CLASS_NAMES)

    print("\n===== FINAL TEST RESULTS =====")
    print(f"Accuracy : {test_acc:.2f}%")
    print(f"Precision: {precision:.4f}")
    print(f"Recall   : {recall:.4f}")
    print(f"F1-score : {f1:.4f}")
    print(report)

    # ================= PLOTS =================
    plt.figure()
    plt.plot(history["train_acc"], label="Train Acc")
    plt.plot(history["val_acc"], label="Val Acc")
    plt.legend(); plt.title("Accuracy")
    plt.savefig("evaluation_results/accuracy_curve.png"); plt.show()

    plt.figure()
    plt.plot(history["train_loss"], label="Train Loss")
    plt.plot(history["val_loss"], label="Val Loss")
    plt.legend(); plt.title("Loss")
    plt.savefig("evaluation_results/loss_curve.png"); plt.show()

    plt.figure(figsize=(6,5))
    plt.imshow(cm, cmap="Blues")
    plt.title("Confusion Matrix")
    plt.colorbar()
    plt.xticks(range(cfg.NUM_CLASSES), cfg.CLASS_NAMES, rotation=45)
    plt.yticks(range(cfg.NUM_CLASSES), cfg.CLASS_NAMES)
    plt.tight_layout()
    plt.savefig("evaluation_results/confusion_matrix.png")
    plt.show()

    return model

# ==========================================================
# RUN
# ==========================================================
model = train_swin()
```

```python
# ==========================================================
# RESNET18 + SWIN TRANSFORMER + CAPSULE NETWORK (FULL PIPELINE)
# JUPYTER + GPU SAFE | WINDOWS SAFE
# ==========================================================

import os, glob, math, json
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from torchvision import transforms, models
import timm

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)

# ==========================================================
# CONFIG
# ==========================================================
class CFG:
    DATASET_PATH = r"D:\Mini\CapsuleNet_Acne"   # 👈 CHANGE IF NEEDED
    CLASSES = ['IGA0', 'IGA1', 'IGA2', 'IGA3', 'IGA4']
    NUM_CLASSES = 5

    IMG_SIZE = 224
    BATCH = 16
    EPOCHS = 20
    LR = 3e-4

    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cfg = CFG()
print("Using device:", cfg.DEVICE)

# ==========================================================
# DATASET
# ==========================================================
class AcneDataset(Dataset):
    def __init__(self, root, split, transform):
        self.paths, self.labels = [], []
        self.transform = transform

        for i, cname in enumerate(cfg.CLASSES):
            folder = os.path.join(root, split, cname)
            if not os.path.exists(folder): continue
            for ext in ("*.jpg", "*.png", "*.jpeg"):
                for p in glob.glob(os.path.join(folder, ext)):
                    self.paths.append(p)
                    self.labels.append(i)

        print(f"✓ {split} set: {len(self.paths)} images")

    def __len__(self): return len(self.paths)

    def __getitem__(self, idx):
        img = Image.open(self.paths[idx]).convert("RGB")
        return self.transform(img), self.labels[idx]

# ==========================================================
# TRANSFORMS
# ==========================================================
train_tf = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.2,0.2,0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

val_tf = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

# ==========================================================
# CAPSULE UTILS
# ==========================================================
def squash(x, dim=-1, eps=1e-7):
    norm = torch.norm(x, dim=dim, keepdim=True)
    return (norm**2 / (1 + norm**2)) * (x / (norm + eps))

class PrimaryCaps(nn.Module):
    def __init__(self, in_dim, num_caps, cap_dim):
        super().__init__()
        self.fc = nn.Linear(in_dim, num_caps * cap_dim)
        self.num_caps, self.cap_dim = num_caps, cap_dim

    def forward(self, x):
        x = self.fc(x)
        x = x.view(x.size(0), self.num_caps, self.cap_dim)
        return squash(x)

class DigitCaps(nn.Module):
    def __init__(self, in_caps, in_dim, out_caps, out_dim, iters=3):
        super().__init__()
        self.iters = iters
        self.W = nn.Parameter(
            0.01 * torch.randn(1, in_caps, out_caps, out_dim, in_dim)
        )

    def forward(self, x):
        B = x.size(0)
        x_hat = torch.matmul(
            self.W.expand(B,-1,-1,-1,-1),
            x.unsqueeze(2).unsqueeze(-1)
        ).squeeze(-1)

        b = torch.zeros(B, x_hat.size(1), x_hat.size(2), 1, device=x.device)

        for i in range(self.iters):
            c = F.softmax(b, dim=2)
            s = (c * x_hat).sum(dim=1)
            v = squash(s)
            if i < self.iters-1:
                b = b + (x_hat * v.unsqueeze(1)).sum(dim=-1, keepdim=True)
        return v

# ==========================================================
# HYBRID MODEL
# ==========================================================
class ResNetSwinCapsNet(nn.Module):
    def __init__(self):
        super().__init__()

        resnet = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        self.resnet = nn.Sequential(*list(resnet.children())[:-1])
        res_dim = 512

        self.swin = timm.create_model(
            "swin_tiny_patch4_window7_224",
            pretrained=True,
            num_classes=0
        )
        swin_dim = self.swin.num_features

        self.fusion = nn.Sequential(
            nn.Linear(res_dim + swin_dim, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        self.primary = PrimaryCaps(512, 32, 8)
        self.digit = DigitCaps(32, 8, cfg.NUM_CLASSES, 16)

    def forward(self, x):
        r = self.resnet(x).flatten(1)
        s = self.swin(x)
        f = self.fusion(torch.cat([r, s], dim=1))
        p = self.primary(f)
        d = self.digit(p)
        logits = torch.norm(d, dim=2)
        return logits

# ==========================================================
# LOSS
# ==========================================================
criterion = nn.CrossEntropyLoss()

# ==========================================================
# TRAIN / EVAL
# ==========================================================
def run_epoch(model, loader, optimizer=None):
    train = optimizer is not None
    model.train() if train else model.eval()

    losses, preds, labels = [], [], []

    for x,y in loader:
        x,y = x.to(cfg.DEVICE), torch.tensor(y).to(cfg.DEVICE)

        if train: optimizer.zero_grad()

        out = model(x)
        loss = criterion(out, y)

        if train:
            loss.backward()
            optimizer.step()

        losses.append(loss.item())
        preds.extend(out.argmax(1).cpu().numpy())
        labels.extend(y.cpu().numpy())

    return (
        np.mean(losses),
        accuracy_score(labels, preds),
        precision_score(labels, preds, average="macro", zero_division=0),
        recall_score(labels, preds, average="macro", zero_division=0),
        f1_score(labels, preds, average="macro", zero_division=0),
        confusion_matrix(labels, preds)
    )

# ==========================================================
# LOAD DATA
# ==========================================================
train_ds = AcneDataset(cfg.DATASET_PATH, "train", train_tf)
val_ds   = AcneDataset(cfg.DATASET_PATH, "valid", val_tf)
test_ds  = AcneDataset(cfg.DATASET_PATH, "test",  val_tf)

train_loader = DataLoader(train_ds, batch_size=cfg.BATCH, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=cfg.BATCH, shuffle=False)
test_loader  = DataLoader(test_ds, batch_size=cfg.BATCH, shuffle=False)

# ==========================================================
# TRAINING
# ==========================================================
model = ResNetSwinCapsNet().to(cfg.DEVICE)
optimizer = optim.AdamW(model.parameters(), lr=cfg.LR)

history = {"train":[], "val":[]}

for epoch in range(cfg.EPOCHS):
    tr = run_epoch(model, train_loader, optimizer)
    va = run_epoch(model, val_loader)

    history["train"].append(tr)
    history["val"].append(va)

    print(f"Epoch {epoch+1}/{cfg.EPOCHS}")
    print(f" Train Acc={tr[1]*100:.2f}% F1={tr[4]:.3f}")
    print(f" Val   Acc={va[1]*100:.2f}% F1={va[4]:.3f}")

# ==========================================================
# TEST + CONFUSION MATRIX
# ==========================================================
loss, acc, prec, rec, f1, cm = run_epoch(model, test_loader)

print("\nTEST RESULTS")
print(f"Accuracy : {acc*100:.2f}%")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")
print(f"F1-score : {f1:.3f}")

print("\nClassification Report")
print(classification_report(
    np.argmax(cm,1), np.argmax(cm,0),
    target_names=cfg.CLASSES, zero_division=0
))

plt.figure(figsize=(6,5))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.colorbar()
plt.xticks(range(cfg.NUM_CLASSES), cfg.CLASSES, rotation=45)
plt.yticks(range(cfg.NUM_CLASSES), cfg.CLASSES)
plt.tight_layout()
plt.show()
```

```python
# ==========================================================
# ✅ RESNET18 + SWIN-T + CAPSULE HEAD (Dynamic Routing)
# ✅ JUPYTER/WINDOWS SAFE (num_workers=0, no multiprocessing pickle issues)
# ✅ GPU (RTX 3060) + AMP autocast fixed (device_type required)
# ✅ Full metrics: Train/Val Loss + Acc + Precision + Recall + F1 (each epoch)
# ✅ Visualizations: curves + CM raw/normalized + sample predictions + capsule length dist
#
# DATASET STRUCTURE (your D drive):
# D:\Mini\CapsuleNet_Acne\
#   train\IGA0, IGA1, IGA2, IGA3, IGA4\
#   valid\IGA0, IGA1, IGA2, IGA3, IGA4\
#   test \IGA0, IGA1, IGA2, IGA3, IGA4\
#
# RUN (in Jupyter):
#   CFG.DATASET_PATH = r"D:\Mini\CapsuleNet_Acne"
#   model, results = train_hybrid_capsnet()
# ==========================================================

import os, glob, json, math, time
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
from torchvision.transforms import RandAugment

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

import matplotlib.pyplot as plt

# AMP (PyTorch 2.x)
from torch.amp import autocast

# ---------------------------
# CONFIG
# ---------------------------
class CFG:
    DATASET_PATH = r"D:\Mini"     # 👈 CHANGE if needed

    CLASS_NAMES = ['IGA0', 'IGA1', 'IGA2', 'IGA3', 'IGA4']
    NUM_CLASSES = 5

    IMG = 224
    BATCH = 16               # RTX 3060: start with 16; increase to 32 if VRAM allows
    EPOCHS = 30
    PATIENCE = 8

    LR = 3e-4
    MIN_LR = 1e-6
    WARMUP = 2
    WD = 1e-4

    # Capsule head (FC primary capsules)
    PRIMARY_CAPS = 256       # number of primary capsules
    PRIMARY_DIM = 8
    DIGIT_DIM = 16
    ROUTING_ITERS = 3

    # Loss
    M_PLUS = 0.9
    M_MINUS = 0.1
    LAMBDA = 0.5
    MARGIN_SCALE = 2.0       # keep modest; backbone is strong
    CE_WARMUP_EPOCHS = 5     # CE on capsule lengths for stabilization

    # Augment
    AUG_START = 1

    # Jupyter/Windows safe
    NUM_WORKERS = 0          # IMPORTANT: avoid multiprocessing pickle issues in notebook
    PIN_MEMORY = True

    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cfg = CFG()

print("Using device:", cfg.DEVICE)
if cfg.DEVICE.type == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))

# ---------------------------
# UTILS
# ---------------------------
def set_seed(seed=42):
    import random
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)

def squash(x, dim=-1, eps=1e-7):
    n = torch.norm(x, p=2, dim=dim, keepdim=True)
    return (n**2 / (1 + n**2)) * (x / (n + eps))

def cosine_lr(epoch):
    if epoch < cfg.WARMUP:
        return cfg.LR * (epoch + 1) / cfg.WARMUP
    t = (epoch - cfg.WARMUP) / max(1, cfg.EPOCHS - cfg.WARMUP)
    return cfg.MIN_LR + 0.5 * (cfg.LR - cfg.MIN_LR) * (1 + math.cos(math.pi * t))

def ce_alpha(epoch):
    return 1.0 if epoch < cfg.CE_WARMUP_EPOCHS else 0.0

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)

def save_json(obj, path):
    with open(path, "w") as f:
        json.dump(obj, f, indent=4)

def to_float32(x):
    return x.float() if x.dtype != torch.float32 else x

# ---------------------------
# DATASET
# ---------------------------
class AcneDataset(Dataset):
    def __init__(self, root, split, transform):
        self.transform = transform
        self.paths = []
        self.labels = []

        split_dir = os.path.join(root, split)
        if not os.path.exists(split_dir):
            print(f"❌ Missing folder: {split_dir}")
            return

        for i, cname in enumerate(cfg.CLASS_NAMES):
            cdir = os.path.join(split_dir, cname)
            if not os.path.exists(cdir):
                print(f"⚠ Missing class folder: {cdir}")
                continue
            for ext in ("*.jpg","*.png","*.jpeg","*.JPG","*.PNG","*.JPEG"):
                self.paths += glob.glob(os.path.join(cdir, ext))
                self.labels += [i] * len(glob.glob(os.path.join(cdir, ext)))

        print(f"✓ {split} set: {len(self.paths)} images")

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, i):
        img = Image.open(self.paths[i]).convert("RGB")
        x = self.transform(img)
        y = self.labels[i]
        return x, y

def get_train_tf(epoch):
    tf = [
        transforms.Resize((cfg.IMG, cfg.IMG)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(12),
        transforms.ColorJitter(0.15, 0.15, 0.10, 0.05),
    ]
    if epoch >= cfg.AUG_START:
        tf.insert(1, RandAugment(num_ops=2, magnitude=9))
    tf += [
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ]
    return transforms.Compose(tf)

val_tf = transforms.Compose([
    transforms.Resize((cfg.IMG, cfg.IMG)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

# ---------------------------
# CAPSULE MODULES
# ---------------------------
class PrimaryCapsFC(nn.Module):
    """
    Projects fused feature vector -> N primary capsules (each of dim D)
    """
    def __init__(self, in_dim, n_caps, cap_dim):
        super().__init__()
        self.n_caps = n_caps
        self.cap_dim = cap_dim
        self.fc = nn.Sequential(
            nn.Linear(in_dim, n_caps * cap_dim),
            nn.ReLU(inplace=True),
        )

    def forward(self, f):
        # f: [B, in_dim]
        B = f.size(0)
        x = self.fc(f).view(B, self.n_caps, self.cap_dim)
        return squash(x, dim=-1)

class DigitCaps(nn.Module):
    def __init__(self, in_caps, in_dim, out_caps, out_dim, iters=3):
        super().__init__()
        self.in_caps = in_caps
        self.out_caps = out_caps
        self.out_dim = out_dim
        self.iters = iters
        self.W = nn.Parameter(0.01 * torch.randn(1, in_caps, out_caps, out_dim, in_dim))

    def forward(self, x):
        # x: [B, in_caps, in_dim]
        B = x.size(0)
        Wb = self.W.expand(B, -1, -1, -1, -1)
        u_hat = torch.matmul(Wb, x.unsqueeze(2).unsqueeze(-1)).squeeze(-1)
        # u_hat: [B, in_caps, out_caps, out_dim]

        b = torch.zeros(B, self.in_caps, self.out_caps, 1, device=x.device, dtype=torch.float32)

        for i in range(self.iters):
            c = F.softmax(b, dim=2)                         # [B, in_caps, out_caps, 1]
            s = (c * u_hat).sum(dim=1)                      # [B, out_caps, out_dim]
            v = squash(s, dim=-1)                           # [B, out_caps, out_dim]
            if i < self.iters - 1:
                b = b + (u_hat * v.unsqueeze(1)).sum(dim=-1, keepdim=True)

        return v

# ---------------------------
# HYBRID MODEL: ResNet18 + Swin-T + Capsules
# ---------------------------
class ResNetSwinCapsNet(nn.Module):
    def __init__(self):
        super().__init__()

        # ResNet18
        res = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        self.resnet = nn.Sequential(*list(res.children())[:-1])  # [B,512,1,1]

        # Swin-T
        swin = models.swin_t(weights=models.Swin_T_Weights.IMAGENET1K_V1)
        self.swin_feat = swin.features
        self.swin_pool = nn.AdaptiveAvgPool2d((1, 1))

        # Infer dims dynamically (avoids your earlier 1280 vs 519 crash)
        with torch.no_grad():
            dummy = torch.zeros(1, 3, cfg.IMG, cfg.IMG)
            r = self.resnet(dummy).flatten(1)
            s = self.swin_pool(self.swin_feat(dummy)).flatten(1)
            self.res_dim = r.shape[1]
            self.swin_dim = s.shape[1]
            fused_dim = self.res_dim + self.swin_dim

        print(f"✓ ResNet dim: {self.res_dim}")
        print(f"✓ Swin dim:   {self.swin_dim}")
        print(f"✓ Fused dim:  {fused_dim}")

        # Fusion
        self.fuse_norm = nn.LayerNorm(fused_dim)
        self.fuse_drop = nn.Dropout(0.2)
        self.gate = nn.Sequential(nn.Linear(fused_dim, fused_dim), nn.Sigmoid())

        # Capsules
        self.primary = PrimaryCapsFC(fused_dim, cfg.PRIMARY_CAPS, cfg.PRIMARY_DIM)
        self.digit = DigitCaps(cfg.PRIMARY_CAPS, cfg.PRIMARY_DIM, cfg.NUM_CLASSES, cfg.DIGIT_DIM, cfg.ROUTING_ITERS)

    def forward(self, x):
        r = self.resnet(x).flatten(1)                        # [B,512]
        s = self.swin_pool(self.swin_feat(x)).flatten(1)     # [B,?]
        f = torch.cat([r, s], dim=1)                         # [B,fused]
        f = self.fuse_norm(f)
        f = self.fuse_drop(f)
        f = f * self.gate(f)

        prim = self.primary(f)                               # [B,PRIMARY_CAPS,PRIMARY_DIM]
        dig = self.digit(prim)                               # [B,NUM_CLASSES,DIGIT_DIM]
        lengths = torch.norm(dig, dim=2)                     # [B,NUM_CLASSES]
        pred = lengths.argmax(dim=1)
        return lengths, pred

# ---------------------------
# LOSS: Margin + CE warmup
# ---------------------------
class CapsuleMarginCELoss(nn.Module):
    def __init__(self, class_counts):
        super().__init__()
        counts = np.asarray(class_counts, dtype=np.float32)
        counts[counts < 1] = 1
        w = counts.sum() / counts
        w = (w / w.mean()).clip(0.6, 2.0).astype(np.float32)
        self.register_buffer("w", torch.tensor(w, dtype=torch.float32))
        self.ce = nn.CrossEntropyLoss(weight=self.w)

    def forward(self, lengths, y, alpha):
        # lengths: [B,C] (float)
        # y: [B] (long)
        y = y.long()
        t = F.one_hot(y, cfg.NUM_CLASSES).float()

        present = F.relu(cfg.M_PLUS - lengths) ** 2
        absent  = F.relu(lengths - cfg.M_MINUS) ** 2
        margin = (t * present + cfg.LAMBDA * (1 - t) * absent).sum(dim=1).mean()

        loss = cfg.MARGIN_SCALE * margin + alpha * self.ce(lengths, y)
        return loss

# ---------------------------
# METRICS
# ---------------------------
@torch.no_grad()
def evaluate_full(model, loader):
    model.eval()
    ys, ps = [], []
    losses = []

    for x, y in loader:
        x = x.to(cfg.DEVICE)
        y = torch.as_tensor(y, device=cfg.DEVICE, dtype=torch.long)

        lengths, pred = model(x)
        ys.extend(y.cpu().numpy().tolist())
        ps.extend(pred.cpu().numpy().tolist())

    ys = np.array(ys)
    ps = np.array(ps)

    acc = accuracy_score(ys, ps) * 100
    prec = precision_score(ys, ps, average="macro", zero_division=0)
    rec  = recall_score(ys, ps, average="macro", zero_division=0)
    f1   = f1_score(ys, ps, average="macro", zero_division=0)

    return acc, prec, rec, f1, ys, ps

# ---------------------------
# VISUALIZATIONS
# ---------------------------
def plot_curves(hist, outdir):
    ensure_dir(outdir)
    epochs = np.arange(1, len(hist["train_acc"]) + 1)

    # Accuracy
    plt.figure(figsize=(7,5))
    plt.plot(epochs, hist["train_acc"], label="Train Acc")
    plt.plot(epochs, hist["val_acc"], label="Val Acc")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy (%)"); plt.title("Accuracy Curve")
    plt.legend(); plt.grid(alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(outdir, "accuracy_curve.png"), dpi=300)
    plt.close()

    # Loss
    plt.figure(figsize=(7,5))
    plt.plot(epochs, hist["train_loss"], label="Train Loss")
    plt.plot(epochs, hist["val_loss"], label="Val Loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss Curve")
    plt.legend(); plt.grid(alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(outdir, "loss_curve.png"), dpi=300)
    plt.close()

    # Precision/Recall/F1
    plt.figure(figsize=(7,5))
    plt.plot(epochs, hist["val_prec"], label="Val Precision")
    plt.plot(epochs, hist["val_rec"], label="Val Recall")
    plt.plot(epochs, hist["val_f1"], label="Val F1")
    plt.xlabel("Epoch"); plt.ylabel("Score"); plt.title("Val Precision / Recall / F1")
    plt.legend(); plt.grid(alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(outdir, "prf_curve.png"), dpi=300)
    plt.close()

    # LR
    plt.figure(figsize=(7,5))
    plt.plot(epochs, hist["lr"])
    plt.xlabel("Epoch"); plt.ylabel("Learning Rate"); plt.title("LR Schedule")
    plt.grid(alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(outdir, "lr_curve.png"), dpi=300)
    plt.close()

def plot_cm(cm, class_names, outpath, normalized=False):
    ensure_dir(os.path.dirname(outpath))
    if normalized:
        cm = cm.astype(np.float32) / (cm.sum(axis=1, keepdims=True) + 1e-6)

    plt.figure(figsize=(7,6))
    plt.imshow(cm, cmap="Blues")
    plt.title("Confusion Matrix" + (" (Normalized)" if normalized else ""))
    plt.colorbar()
    plt.xticks(range(len(class_names)), class_names, rotation=45, ha="right")
    plt.yticks(range(len(class_names)), class_names)

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            txt = f"{cm[i,j]*100:.1f}%" if normalized else str(int(cm[i,j]))
            plt.text(j, i, txt, ha="center", va="center",
                     color="white" if cm[i,j] > (0.5 if normalized else cm.max()/2) else "black",
                     fontsize=9)

    plt.tight_layout()
    plt.savefig(outpath, dpi=300)
    plt.close()

@torch.no_grad()
def visualize_predictions(model, loader, outpath, max_images=12):
    ensure_dir(os.path.dirname(outpath))
    model.eval()

    # grab one batch
    for x, y in loader:
        x = x.to(cfg.DEVICE)
        lengths, pred = model(x)
        pred = pred.cpu().numpy()
        y = np.array(y)

        n = min(max_images, x.size(0))
        cols = 4
        rows = int(math.ceil(n / cols))

        plt.figure(figsize=(12, 3*rows))
        for i in range(n):
            img = x[i].detach().cpu().permute(1,2,0).numpy()
            # de-normalize for nicer display (approx)
            img = (img - img.min()) / (img.max() - img.min() + 1e-6)

            gt = cfg.CLASS_NAMES[y[i]]
            pr = cfg.CLASS_NAMES[pred[i]]

            plt.subplot(rows, cols, i+1)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"GT: {gt} | Pred: {pr}", color=("green" if gt==pr else "red"), fontsize=10)

        plt.tight_layout()
        plt.savefig(outpath, dpi=300)
        plt.close()
        return

# ---------------------------
# GPU MONITORING (optional)
# ---------------------------
def print_gpu_status():
    if cfg.DEVICE.type != "cuda":
        print("CUDA not available.")
        return
    alloc = torch.cuda.memory_allocated() / (1024**2)
    reserv = torch.cuda.memory_reserved() / (1024**2)
    print(f"GPU mem allocated: {alloc:.1f} MiB | reserved: {reserv:.1f} MiB")

# ---------------------------
# TRAINING
# ---------------------------
def train_hybrid_capsnet():
    out_eval = os.path.join(cfg.DATASET_PATH, "evaluation_results")
    out_models = os.path.join(cfg.DATASET_PATH, "saved_models")
    ensure_dir(out_eval); ensure_dir(out_models)

    # Datasets
    train_ds = AcneDataset(cfg.DATASET_PATH, "train", get_train_tf(epoch=0))
    val_ds   = AcneDataset(cfg.DATASET_PATH, "valid", val_tf)
    test_ds  = AcneDataset(cfg.DATASET_PATH, "test",  val_tf)

    if len(train_ds) == 0 or len(val_ds) == 0 or len(test_ds) == 0:
        raise ValueError("One of the splits has 0 images. Check folder names: train/valid/test and class folders.")

    # Class counts + sampler
    counts = np.bincount(train_ds.labels, minlength=cfg.NUM_CLASSES)
    print("✓ Class counts (train):", counts.tolist())

    sample_w = 1.0 / (counts[np.array(train_ds.labels)] + 1e-6)
    sampler = WeightedRandomSampler(sample_w, num_samples=len(sample_w), replacement=True)

    # Loaders (Windows/Jupyter: num_workers=0)
    train_loader = DataLoader(train_ds, batch_size=cfg.BATCH, sampler=sampler,
                              num_workers=cfg.NUM_WORKERS, pin_memory=cfg.PIN_MEMORY)
    val_loader   = DataLoader(val_ds, batch_size=cfg.BATCH, shuffle=False,
                              num_workers=cfg.NUM_WORKERS, pin_memory=cfg.PIN_MEMORY)
    test_loader  = DataLoader(test_ds, batch_size=cfg.BATCH, shuffle=False,
                              num_workers=cfg.NUM_WORKERS, pin_memory=cfg.PIN_MEMORY)

    # Model
    model = ResNetSwinCapsNet().to(cfg.DEVICE)

    # Loss + optimizer
    loss_fn = CapsuleMarginCELoss(counts).to(cfg.DEVICE)
    opt = optim.AdamW(model.parameters(), lr=cfg.LR, weight_decay=cfg.WD)

    # GradScaler (robust across torch versions)
    try:
        from torch.amp import GradScaler
        scaler = GradScaler("cuda", enabled=(cfg.DEVICE.type == "cuda"))
    except Exception:
        from torch.cuda.amp import GradScaler
        scaler = GradScaler(enabled=(cfg.DEVICE.type == "cuda"))

    hist = {
        "train_loss": [], "val_loss": [],
        "train_acc": [], "val_acc": [],
        "val_prec": [], "val_rec": [], "val_f1": [],
        "lr": []
    }

    best_val = 0.0
    best_path = os.path.join(out_models, "best_resnet_swin_caps.pth")
    patience = 0

    for epoch in range(cfg.EPOCHS):
        t0 = time.time()
        model.train()
        train_ds.transform = get_train_tf(epoch)

        lr = cosine_lr(epoch)
        for g in opt.param_groups:
            g["lr"] = lr

        alpha = ce_alpha(epoch)

        # Train
        running_loss = 0.0
        correct = 0
        total = 0

        for x, y in train_loader:
            x = x.to(cfg.DEVICE, non_blocking=True)
            y = torch.as_tensor(y, device=cfg.DEVICE, dtype=torch.long)

            opt.zero_grad(set_to_none=True)

            with autocast(device_type=cfg.DEVICE.type, enabled=(cfg.DEVICE.type == "cuda")):
                lengths, pred = model(x)
                lengths = to_float32(lengths)  # ensure float32 for CE
                loss = loss_fn(lengths, y, alpha)

            scaler.scale(loss).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt)
            scaler.update()

            running_loss += float(loss.item())
            correct += int((pred == y).sum().item())
            total += int(y.size(0))

        train_loss = running_loss / max(1, len(train_loader))
        train_acc = 100.0 * correct / max(1, total)

        # Val (compute val loss too)
        model.eval()
        vloss_sum = 0.0
        with torch.no_grad():
            for x, y in val_loader:
                x = x.to(cfg.DEVICE)
                y = torch.as_tensor(y, device=cfg.DEVICE, dtype=torch.long)
                with autocast(device_type=cfg.DEVICE.type, enabled=(cfg.DEVICE.type == "cuda")):
                    lengths, pred = model(x)
                    lengths = to_float32(lengths)
                    vloss = loss_fn(lengths, y, alpha=0.0)  # CE warmup not needed for val loss display
                vloss_sum += float(vloss.item())

        val_loss = vloss_sum / max(1, len(val_loader))
        val_acc, val_prec, val_rec, val_f1, _, _ = evaluate_full(model, val_loader)

        # Log
        hist["train_loss"].append(train_loss)
        hist["val_loss"].append(val_loss)
        hist["train_acc"].append(train_acc)
        hist["val_acc"].append(val_acc)
        hist["val_prec"].append(val_prec)
        hist["val_rec"].append(val_rec)
        hist["val_f1"].append(val_f1)
        hist["lr"].append(lr)

        dt = time.time() - t0
        print(f"Epoch {epoch+1:02d}/{cfg.EPOCHS} | {dt:.1f}s | LR={lr:.2e} | CEα={alpha:.2f}")
        print(f"  Train: loss={train_loss:.4f}, acc={train_acc:.2f}%")
        print(f"  Val:   loss={val_loss:.4f}, acc={val_acc:.2f}% | P={val_prec:.3f} R={val_rec:.3f} F1={val_f1:.3f}")
        if cfg.DEVICE.type == "cuda":
            print_gpu_status()

        # Early stopping
        if val_acc > best_val:
            best_val = val_acc
            patience = 0
            torch.save({
                "epoch": epoch+1,
                "model": model.state_dict(),
                "best_val": best_val,
                "cfg": vars(cfg)
            }, best_path)
            print(f"  ✓ Best saved: {best_val:.2f}%")
        else:
            patience += 1
            if patience >= cfg.PATIENCE:
                print(f"⏹ Early stopping (patience={cfg.PATIENCE})")
                break

    # Load best
    ckpt = torch.load(best_path, map_location=cfg.DEVICE)
    model.load_state_dict(ckpt["model"])

    # Test
    test_acc, test_prec, test_rec, test_f1, y_true, y_pred = evaluate_full(model, test_loader)
    cm = confusion_matrix(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=cfg.CLASS_NAMES, zero_division=0)

    results = {
        "best_val_acc": float(best_val),
        "test_acc": float(test_acc),
        "test_macro_precision": float(test_prec),
        "test_macro_recall": float(test_rec),
        "test_macro_f1": float(test_f1),
        "confusion_matrix": cm.tolist(),
        "class_names": cfg.CLASS_NAMES,
        "report": report,
        "history": hist
    }

    # Save
    save_json(results, os.path.join(out_eval, "results.json"))
    with open(os.path.join(out_eval, "classification_report.txt"), "w") as f:
        f.write(report)

    # Plots
    plot_curves(hist, out_eval)
    plot_cm(cm, cfg.CLASS_NAMES, os.path.join(out_eval, "confusion_matrix.png"), normalized=False)
    plot_cm(cm, cfg.CLASS_NAMES, os.path.join(out_eval, "confusion_matrix_normalized.png"), normalized=True)
    visualize_predictions(model, test_loader, os.path.join(out_eval, "sample_predictions.png"), max_images=12)

    print("\n" + "="*60)
    print("FINAL (BEST MODEL)")
    print("="*60)
    print(f"Best Val Acc: {best_val:.2f}%")
    print(f"Test Acc:     {test_acc:.2f}%")
    print(f"Test P/R/F1:  {test_prec:.3f} / {test_rec:.3f} / {test_f1:.3f}")
    print("\nClassification Report:\n", report)
    print("\nSaved outputs to:")
    print("  ", out_eval)
    print("  ", out_models)

    return model, results


# ==========================================================
# RUN
# ==========================================================
# Make sure your dataset path is correct:
# cfg.DATASET_PATH = r"D:\Mini\CapsuleNet_Acne"
model, results = train_hybrid_capsnet()

# ==========================================================
# OPTIONAL: GPU UTILIZATION (Windows)
# In a separate CMD:
#   nvidia-smi -l 1
# ==========================================================
```
